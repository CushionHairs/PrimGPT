```python
import json

def convert_cookies_to_netscape(json_file, netscape_file):
    with open(json_file, 'r') as f:
        cookies = json.load(f)

    with open(netscape_file, 'w') as f:
        f.write("# Netscape HTTP Cookie File\n")
        f.write("# This file was generated by a script\n")
        f.write("# Edit at your own risk.\n")
        f.write("\n")

        for cookie in cookies:
            domain = cookie['domain']
            path = cookie['path']
            secure = 'TRUE' if cookie['secure'] else 'FALSE'
            http_only = 'TRUE' if cookie.get('httpOnly', False) else ''
            expiration_date = int(cookie['expirationDate'])
            name = cookie['name']
            value = cookie['value']

            # Write the cookie in Netscape format
            f.write(f"{domain}\tTRUE\t{path}\t{expiration_date}\t{name}\t{value}\n")

# 사용 예시
convert_cookies_to_netscape('yt-cookies.txt', 'yt-cookies.netscape.txt')
```
